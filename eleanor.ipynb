```python
from IPython.display import Image

import warnings
warnings.filterwarnings('ignore')
import eleanor
import numpy as np
from astropy import units as u
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
```

```python
eleanor.Update(sector=17)
```

```python
star_5614834 = eleanor.Source(tic=5614834, sector=17)
```

```python
data_5614834 = eleanor.TargetData(star_5614834, height=15, width=15, bkg_size=31, do_psf=True, do_pca=True, regressors='corner')
```

```python
plt.figure(figsize=(15,5))

q = data_5614834.quality == 0

# long term trend - plausible? systematic?
# partial helium ionisation
# wiki instability strip
# pca_flux removes the long-term trend
plt.plot(data_5614834.time[q], data_5614834.pca_flux[q]/np.nanmedian(data_5614834.pca_flux[q]), '.')

# raw_flux - removes background and nothing else
# corr_flux - tries to get as flat a light curve as possible. Regresses against anything that could be instrumental (centroids, time, background, co-training basis vectors (things that are shared between all the other stars in the detector))
# pca_flux - takes the raw flux and only regresses against basis vectors - remove anything that is clearly shared w/ other stars

plt.ylabel('Normalized Flux')
plt.title('TIC 5614834');
plt.ylim(0.9, 1.2)
```
